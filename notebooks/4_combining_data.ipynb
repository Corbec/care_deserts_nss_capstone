```python
import pandas as pd
import numpy as np
```

```python
total_pop = pd.read_csv('../data/raw_census/B01003.csv')

total_pop
```

```python
total_pop = total_pop.drop(['NAME', 'GEO_ID'], axis = 1)
```

```python
total_pop = total_pop[['zip code tabulation area', 'state', 'B01003_001E']]
```

```python
total_pop = total_pop.rename(columns = {'zip code tabulation area': 'zip', 'B01003_001E': 'total_pop'})
```

```python
total_pop.to_csv('../data/total_pop.csv',
        index = False)
```

```python
provider_zip = pd.read_csv('../data/rendering_zipcode.csv')
```

```python
provider_zip
```

```python
provider_zip = provider_zip.merge(total_pop,
                                how = 'inner',
                                left_on = 'rendering_zip',
                                right_on = 'zip')
```

```python
provider_zip = provider_zip.drop('zip', axis = 1)
```

```python
provider_zip['provider_per_1000'] = provider_zip['rendering_npi'] / (provider_zip['total_pop'] / 1000)
```

```python
provider_zip
```

```python
provider_zip[provider_zip['provider_per_1000'] == provider_zip['provider_per_1000'].max()]
```

```python
provider_zip = provider_zip[provider_zip['total_pop'] != 0]
```

```python
provider_zip[provider_zip['provider_per_1000'] == provider_zip['provider_per_1000'].max()]
```

```python
median_income = pd.read_csv('../data/median_income.csv')

median_income = median_income.drop(['NAME', 'GEO_ID'], axis = 1)

median_income = median_income[['zip code tabulation area', 'state', 'B19013_001E']].rename(columns = {'zip code tabulation area': 'zip', 'B19013_001E': 'median_income'})

median_income
```

```python
provider_zip = provider_zip.merge(median_income,
                                how = 'inner',
                                left_on = 'rendering_zip',
                                right_on = 'zip').drop('zip', axis = 1)
```

```python
provider_zip
```

```python
provider_zip.to_csv('../data/provider_zip_pop.csv',
        index = False)
```
