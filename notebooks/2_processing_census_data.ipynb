```python
def process_dataframes(df, csv_name):
     
     df1 = df[df['state'] == '47']

     df2 = df1[df1.columns.drop(list(df1.filter(regex='\w+\d+.*PM$|\w+\d+.M$|\w+\d+.PE$|\w+\d+.PEA$|\w+\d+.*MA$|\w+\d+.*EA$|\w+\d+.*PMA$')))]

     df2.to_csv('../data/' + csv_name,
               index = False)

     return df2
```
