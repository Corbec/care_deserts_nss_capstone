```python
with open('../data/census_dict.pkl', 'rb') as f:
    census_dict = pickle.load(f)
```

```python
# adding table on vehicle ownership to dictionary
with open('../data/census_key.json') as fi:
    credentials = json.load(fi)
    
api_key = credentials['api_key']

c = Census(api_key)

no_cars_2020 = c.acs5.zipcode(('NAME', 'group(B08201)'), zcta = Census.ALL, state = 47)

census_dict['group(B08201)'] = no_cars_2020
```
