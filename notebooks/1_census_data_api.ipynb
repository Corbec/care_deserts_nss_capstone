```python
census_tables = {
    'B25045_001E': 5,
    'B25046_001E': 5,
    'DP05_0001E': 1,
    'S0102_C01_001E': 1,
    'S0103_C01_001E': 1,
    'S1701_C01_001E': 1,
    'S1703_C01_001E': 1,
    'S1810_C01_001E': 1,
    'S1811_C01_001E': 1,
    'S1903_C01_001E': 1,
    'S2701_C01_001E': 1,
    'S2702_C01_001E': 1,
    'B01003_001E': 5,
    'B02001_001E': 5,
    'B09010_001E': 5,
    'B09020_001E': 5
}
```

```python
def get_census_data(table, num):
    if num == 5:
        table = c.acs5.zipcode(('NAME', table), state = states.TN.fips, zcta = Census.ALL)
    if num == 1:
        table = c.acs1.zipcode(('NAME', table), state = states.TN.fips, zcta = Census.ALL)
    
    return table
```

```python
census_dict = {}

for table, acs in census_tables.items():
    census_dict[table] = get_census_data(table, acs)
```
