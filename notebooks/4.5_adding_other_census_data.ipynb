```python
vehicles = pd.read_csv('../data/raw_census/B08201.csv')
```

```python
vehicles
```

```python
no_vehicles = vehicles[['zip code tabulation area',
                    'state',
                    'B08201_001E',
                    'B08201_002E',
                    'B08201_007E',
                    'B08201_008E']].rename(columns = {'zip code tabulation area': 'zip',
                                                     'B08201_001E': 'total_hh',
                                                     'B08201_002E': 'total_no_veh',
                                                     'B08201_007E': 'total_1_person_hh',
                                                     'B08201_008E': '1_person_hh_no_veh'})
```

```python
no_vehicles['total_no_car_ratio'] = no_vehicles['total_no_veh'] / no_vehicles['total_hh']

no_vehicles['1_per_no_car_ratio'] = no_vehicles['1_person_hh_no_veh'] / no_vehicles['total_1_person_hh']
```

```python
no_vehicles
```

```python
no_vehicles[no_vehicles['total_no_car_ratio'] == no_vehicles['total_no_car_ratio'].max()]
```

```python
no_vehicles[no_vehicles['total_1_person_hh'] > 100].max()
```

```python
provider_zip = provider_zip.merge(no_vehicles,
                                how = 'left',
                                left_on = 'rendering_zip',
                                right_on = 'zip').drop('zip', axis = 1)
```

```python
provider_zip
```

```python
provider_zip.to_csv('../data/provider_zip_pop_cars.csv', index = False)
```
